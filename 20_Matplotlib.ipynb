## 제목과 레이블, 스타일
* 제목은 plt.title(제목)
* plt.xlabel(), plt.ylzbel() 로 x 축, y 축 에 이름을 붙일수있음
* plot() 안에 label 옵션을 이용하면 해당 그래프 의 이름도 정의할수있음
* label 인자를 따로 주지않으면 legend() 는 아무것도 출력하지않을것이다.
* legend() 에 옵션을 주어 범례의 위치를 표시할수있다.
 * 옵션은 best, upper right, lower center 등이있음

```python
x = np.linspace(0, np.pi * 2, 100)

# plt.style.use("Solarize_Light2")
plt.style.use("default")
plt.title("Sin cos curve")  # 제목 설정
plt.plot(x, np.sin(x), "r-", label="sin curve")
plt.plot(x, np.cos(x), "b:", label="cos curve")
plt.xlabel("x value")
plt.ylabel("y value")
plt.legend(loc="best")  # plot() 에서 준 label 을 기반으로 범례 추가
plt.show()
```

* plt.style.available 로 사용가능한 스타일이 나오며 
* plt.style.use("_mpl-gallery") 로 스타일을 적용할수있음

## Scatter()

```python
x = np.arange(0, 10)
plt.axis([0, 10, 0, 100])
plt.scatter(x, x**2, c="red", s=10)  # s 를 이용해 찍히는 점의 크기
plt.text(3, 50, "y=x^2 graph")  # 임의로 원하는 좌표에 텍스트 추가하기
plt.show()
```

## 어노테이션 사용하기
* plt.anntate('표시할 문자열', xy,xyyext,arrowprops) 식으로 사용

```python
x = np.arange(0, 10)
plt.axis([0, 10, 0, 100])
plt.scatter(x, x**2, c="red", s=10)  # s 를 이용해 찍히는 점의 크기
plt.text(3, 50, "y=x^2 graph")  # 임의로 원하는 좌표에 텍스트 추가하기
plt.annotate("annotate", xy=(4, 16), xytext=(8, 40), arrowprops={"color": "red"})
plt.show()
```

## plt.ticks() 사용하기
* 그래프에 표시될 x, y 축의 범위를 직접 지정하기

```python
import matplotlib.pyplot as plt
import numpy as np

# 한글폰트를 사용하기 위한 작업
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

x = np.linspace(-4, 4, 100)
sinY = np.sin(x)
cosY = np.cos(x)
plt.plot(x, sinY, x, cosY)
plt.xticks([-4, -2, 0, 2, 4])
plt.yticks([-1, 0, 1])
plt.title("삼각함수")
```

```python
plt.subplot(211).set(xticks=[-4, -2, 0, 2, 4], yticks=[-1, 0, 1])
triXAxis = np.linspace(-4, 4, 100)
sinYAxis = np.sin(triXAxis)
cosYAxis = np.cos(triXAxis)
plt.plot(triXAxis, sinYAxis)
plt.plot(triXAxis, cosYAxis)
plt.title("Trigonomtric Functions")

# lim=[] 범위를 지정할수있음
plt.subplot(212).set(xlim=[0, 8], ylim=[0, 3])
logXAxis = np.linspace(1, 10, 100)
logYAxis = np.log(logXAxis)
plt.plot(logXAxis, logYAxis)
plt.title("Log Function")
```

## bar() 로 막대그래프 만들기

```python
x = np.arange(3)
years = ["2010", "2011", "2012"]
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]

plt.bar(x, domestic, label="domestic")
plt.bar(x, foreign, label="foregin")
plt.xticks(x, years)
plt.legend()
```

## barh() 를 이용하여 수평막대를 그릴수있음

```python
x = np.arange(3)
years = ["2010", "2011", "2012"]
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]

plt.barh(x, domestic, height=0.3, label="domestic")
plt.barh(x - 0.4, foreign, height=0.3, label="foregin")
plt.yticks(x, years)
plt.legend(loc="lower right")
```

## 파이차트 plt.pie()
* autopct = 표시하는 숫자의 포맷지정
* colors = 색상지정

```python
data = [5, 4, 6, 11]
clist = ["cyan", "gray", "orange", "red"]
plt.pie(data, autopct="%.2f", colors=clist, explode=[0.06, 0.07, 0.08, 0.09])
```
