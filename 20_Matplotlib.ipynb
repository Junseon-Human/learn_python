```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot(
    [1, 2, 3, 4]
)  # x,y 두 값을 주지않으면 기본적으로 y 값이 들어감 x 값을 주지않으면 [0,1,...n] 까지 만들어짐
plt.ylabel("y label")
plt.xlabel("x label")
plt.show()
```

```python
import numpy as np

x = np.arange(10)
plt.plot(x**2)
plt.axis([0, 100, 0, 100])  # x 의 범위 0 ~ 100 y의 범위 0 ~ 100
plt.show()
```

```python
x = np.arange(-20, 20)
y1 = 2 * x
y2 = (1 / 3) * x**2 + 5
y3 = -(x**2) - 5
plt.plot(x, y1, "g--", x, y2, "r^", x, y3, "b*:")
plt.axis([-30, 30, -30, 30])
plt.show()
```
